```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
def normal_x(X, Y):
    X1 = sm.add_constant(X)
    result = sm.OLS(Y, X1).fit()
    print("coeff: ", result.params)
    print('Confidence interval:\n', result.conf_int(alpha=0.05, cols=None))

def x_squared(X, Y):
    X2 = np.transpose(np.vstack((X, X**2)))
    X2 = sm.add_constant(X2)
    result = sm.OLS(Y, X2).fit()
    print("model y ~ x + x^2", result.params)

def two_vars(x1, x2, y):
    # fig, ax = plt.subplots()
    # plt.scatter(x1, x2)
    X = np.transpose(np.vstack((x1, x2)))
    X = sm.add_constant(X)
    print('model y ~ x1 + x2')
    result = sm.OLS(y, X).fit()
    print(result.summary2())
    print('coeff: ', result.params)
    print('p-values', result.pvalues)
    print('model y ~ x1')
    result = sm.OLS(y, X[:, [0,1]]).fit()
    print('coeff: ', result.params)
    print('p-values', result.pvalues)
    print('model y ~ x2')
    result = sm.OLS(y, X[:, [0,2]]).fit()
    print('coeff: ', result.params)
    print('p-values', result.pvalues)
    print('\n')
```

```python
np.random.seed(0)
print('\n####### Ex.13 #######\n')
X = np.random.normal(0.0, 1.0, 100)
Y_normal = -1 + 0.5*X + np.random.normal(0.0, 0.25, 100)
Y_noisy = -1 + 0.5*X + np.random.normal(0.0, 0.5, 100)
Y_less_noisy = -1 + 0.5*X + np.random.normal(0.0, 0.15, 100)
normal_x(X, Y_normal)
normal_x(X, Y_noisy)
normal_x(X, Y_less_noisy)
x_squared(X, Y_normal)
```

```python
np.random.seed(0)
print('\n####### Ex.14 #######\n')
x1 = np.random.rand(100)
x2 = 0.5*x1 + np.random.normal(0.5, 1.0, 100) / 10
y = 2 + 2*x1 + 0.3*x2 + np.random.normal(0.5, 1.0, 100)
two_vars(x1, x2, y)
# after addition of one outlier measurement:
x1 = np.append(x1, [0.1])
x2 = np.append(x2, [0.8])
y = np.append(y, [6.])
two_vars(x1, x2, y)
```
